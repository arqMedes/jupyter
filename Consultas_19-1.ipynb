```python
def carregarHorarioCSV(nomeArq="Horario19_1_1.csv"):
    H2019_1 = pd.read_csv(nomeArq, delimiter=';', index_col=0)
    #, names=['Seg', 'Ter', 'Qua','Qui','Sex')]
    return H2019_1

def substituirProf(H2019_1, profAntigo, profNovo, naTurma = 'Todas'):
    
    if (naTurma == 'Todas'):
        myTurma = H2019_1
    else:
        filtro = H2019_1['Turma'] == naTurma
        myTurma = H2019_1[filtro] 
        
    filtro = myTurma.SEG.str.contains(profAntigo) | myTurma.TER.str.contains(profAntigo) 
    filtro = filtro | myTurma.QUA.str.contains(profAntigo) | myTurma.QUI.str.contains(profAntigo)
    filtro = filtro | myTurma.SEX.str.contains(profAntigo) | myTurma.SAB.str.contains(profAntigo)
    retornando = myTurma[filtro]
    for row in retornando.itertuples():
        i = getattr(row, 'Index')
        #turma = getattr(row, 'Turma')
        seg = getattr(row, 'SEG')
        ter = getattr(row, 'TER')
        qua = getattr(row, 'QUA')
        qui = getattr(row, 'QUI')
        sex = getattr(row, 'SEX')
        sab = getattr(row, 'SAB')
        
        if(seg.__contains__(profAntigo)):
            H2019_1.loc[[i],['SEG']] = seg.replace(profAntigo, profNovo)
        if(ter.__contains__(profAntigo)):
            H2019_1.loc[[i],['TER']] = ter.replace(profAntigo, profNovo)
        if(qua.__contains__(profAntigo)):
            H2019_1.loc[[i],['QUA']] = qua.replace(profAntigo, profNovo)
        if(qui.__contains__(profAntigo)):
            H2019_1.loc[[i],['QUI']] = qui.replace(profAntigo, profNovo)
        if(sex.__contains__(profAntigo)):
            H2019_1.loc[[i],['SEX']] = sex.replace(profAntigo, profNovo)
        if(sab.__contains__(profAntigo)):
            H2019_1.loc[[i],['SAB']] = sab.replace(profAntigo, profNovo)
    print ('ALTERADO\n')
    return



def aulasProf(Hfinal, prof, cab=1):
    
    Hfinal.fillna('',  inplace=True)
    lst = []
    filtro = Hfinal.SEG.str.contains(prof) | Hfinal.TER.str.contains(prof) 
    filtro = filtro | Hfinal.QUA.str.contains(prof) | Hfinal.QUI.str.contains(prof)
    filtro = filtro | Hfinal.SEX.str.contains(prof) | Hfinal.SAB.str.contains(prof)
    retornando = Hfinal[filtro]
    #cont = 0
    for row in retornando.itertuples():
        turma = getattr(row, 'Turma')
        horas = getattr(row, 'index')
        seg = getattr(row, 'SEG')
        ter = getattr(row, 'TER')
        qua = getattr(row, 'QUA')
        qui = getattr(row, 'QUI')
        sex = getattr(row, 'SEX')
        sab = getattr(row, 'SAB')
        
        if(seg.__contains__(prof)):
            valores =seg.split(' | ')
            disciplina = valores[0]
            prof = valores[1]
            sala= valores[2]
               
            mD = (turma , 'SEG', horas , disciplina, sala)
            lst.append(mD)
            
        if(ter.__contains__(prof)):
            valores =ter.split(' | ')
            disciplina = valores[0]
            prof = valores[1]
            sala= valores[2]
               
            mD = (turma , 'TER', horas , disciplina, sala)
            lst.append(mD)
        if(qua.__contains__(prof)):
            valores =qua.split(' | ')
            disciplina = valores[0]
            prof = valores[1]
            sala= valores[2]
               
            mD = (turma , 'QUA', horas , disciplina, sala)
            lst.append(mD)
        if(qui.__contains__(prof)):
            valores =qui.split(' | ')
            disciplina = valores[0]
            prof = valores[1]
            sala= valores[2]
               
            mD = (turma , 'QUI', horas , disciplina + ' | ' + sala)
            lst.append(mD)
        if(sex.__contains__(prof)):
            valores =sex.split(' | ')
            disciplina = valores[0]
            prof = valores[1]
            sala= valores[2]
               
            mD = (turma , 'SEX', horas , disciplina , sala)
            lst.append(mD)
        if(sab.__contains__(prof)):
            valores =sab.split(' | ')
            disciplina = valores[0]
            prof = valores[1]
            sala= valores[2]
               
            mD = (turma , 'SAB', horas , disciplina , sala)
            lst.append(mD)
    if cab:
        print (90*'*')
        print ("                      Prof.: %s " %prof)
        print (90*'*')
    hProf = pd.DataFrame(lst,columns=['Turma','Dia','Aula','Disciplina', 'Sala'])
    return hProf

def quantAulas(H, profs):
    lst = []
    for prof in profs:
        hProf = aulasProf(H, prof.upper(), 0)
        lst.append((prof, hProf.shape[0]))
    return lst
```
