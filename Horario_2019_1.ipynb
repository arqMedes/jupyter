## Agora, 2019-1

```python
def disciplinas(curso = 'ELN', periodo = 4):
    turma = curso + str(periodo)
    if curso ==  'EE':
        turma += 'N'
        url="https://raw.githubusercontent.com/arqMedes/Data/master/matriz_EE.csv"
    if curso ==  'ELN':   
        turma += 'M'
        url="https://raw.githubusercontent.com/arqMedes/Data/master/mCTIELN.csv"
    
    
    matriz=pd.read_csv(url, delimiter =';')
    matriz = matriz[matriz.PERIODO == periodo]
    
    lista1 = list(matriz.loc[:,'COMP_CURRICULAR'])
    lista2 = list(matriz.loc[:,'CREDITOS'])
    return turma, lista1, lista2

def novaTurma19_1(turma, listaDisciplinas, listaHorario):
    dados = {'Seg': ['1', '2', '3'],
             'Ter': ['1', '2', '3'],
             'Qua': ['1', '2', '3'],
             'Qui': ['1', '2', '3'],
             'Sex': ['1', '2', '3']
        }
    if turma[-1] == 'N':
        dados = {'Seg': ['1', '2', '3'],
                 'Ter': ['1', '2', '3'],
                 'Qua': ['1', '2', '3'],
                 'Qui': ['1', '2', '3'],
                 'Sex': ['1', '2', '3'],
                 'Sab': ['1', '2', '3']
        }

    h2019 = pd.DataFrame(dados, index = ['1ºH', '2ºH', '3ºH'])
    t = len(listaHorario)
    lista = range(t)
    for i in lista:
        dh = listaHorario[i]
        for h in dh:
            horas = str(h[1]) + 'ºH'
            if (h[0] == 2):
                h2019.loc[[horas], ['Seg']]= listaDisciplinas[i][:10]
            elif (h[0] == 3):
                 h2019.loc[[horas], ['Ter']]= listaDisciplinas[i][:10]
            elif (h[0] == 4):
                 h2019.loc[[horas], ['Qua']]= listaDisciplinas[i][:10]
            elif (h[0] == 5):
                 h2019.loc[[horas], ['Qui']]= listaDisciplinas[i][:10]
            elif (h[0] == 6):
                 h2019.loc[[horas], ['Sex']]= listaDisciplinas[i][:10]
    
    return h2019
   
```

```python
t, disc, ch = disciplinas('ELN', 4)
#t, disc, ch
#novaTurma19_1M('ELN', 4)
```
