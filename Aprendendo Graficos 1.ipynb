```python
def f(x):
    return x**3
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)
y = f(x)
plt.plot(x,y)
```

```python
def g(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
```

```python
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.plot(t1, g(t1), 'bo', t2, g(t2), 'k')
```
